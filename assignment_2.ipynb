```python
English = float(input('Enter english score'))
Maths = float(input('enter maths score'))
Science = float(input('enter science score'))
average_score = (English + Maths + Science)/3
if average_score == 0 :
    print('Oh No!, You failed everything')
if average_score == 100:
    print('Wow,You got a perfect score')
elif average_score >= 90:
    print('Excellent! You got an A')
elif average_score >= 80:
    print('Very Good! You got a B')
elif average_score >= 70:
    print('Good! You got a C')
elif average_score >= 60:
    print('Needs Improvement! You got a D')
elif average_score <60 :
    print('Fail! You got a F')
else:
    print('Invalid input ,Try again')
```
